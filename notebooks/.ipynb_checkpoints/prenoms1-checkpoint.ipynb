```python
import pandas
import numpy as np
from matplotlib import pyplot as plt

from ipywidgets import interact
```

```python
df = pandas.read_csv('../data/dpt2018.csv', sep=';', encoding='utf8', na_values=['XXXX', 'XX'])
```

### Proportion h/f

```python
col1 = df.groupby(['sexe', 'annais', 'preusuel']).nombre.sum()
col1 = col1.reset_index()
```

```python
tmp = col1.groupby(['annais', 'sexe']).nombre.sum()
annee = col1.annais.unique()
nbh = tmp.values[::2]
nbf = tmp.values[1::2]

plt.plot(annee, nbh / (nbh + nbf))
plt.axhline(0.5, ls='--', c='gray')
plt.xlabel('annee')
plt.ylabel('prop h')
plt.show()
```

### prop marie 1900

```python
nb1900 = col1[col1.annais == 1900].nombre.sum()
nbmarie1900 = col1[(col1.annais == 1900) & (col1.preusuel == 'MARIE')].nombre.sum()
print('prop all maries 1900', nbmarie1900 / nb1900)
print('prop f maries 1900', col1[(col1.annais == 1900)].nombre.max() / nb1900)
```

### nbr prenoms

```python
col1.groupby('annais').preusuel.nunique().plot()
plt.ylabel('nbr premons')
plt.show()
```

```python
plt.plot(annee, col1.groupby('preusuel').annais.min().value_counts().sort_index().cumsum())
plt.xlabel('annee')
plt.ylabel('nbr premons cumulated')
plt.show()
```

```python
df_map = pandas.read_csv('../data/EUCircos_Regions_departements_circonscriptions_communes_gps.csv', sep=';')
```

For each departements: get gps adress of prefecture

```python
df_map.columns
```

```python
prefs = df_map.groupby('prfecture').numro_dpartement.unique()

#print(sorted(prefs))
df_adress = df_map.loc[df_map.nom_commune.isin(prefs.keys())]
df_adress = df_adress.loc[df_adress.numro_circonscription == 1]
df_adress = df_adress.loc[df_adress.apply(lambda x: x.numro_dpartement == prefs[x.nom_commune][0], axis=1)]

# replace corse
tmp = np.mean(df_adress.loc[df_adress.numro_dpartement.isin(['2A', '2B']), 
                            ['latitude', 'longitude']].astype(float))

df_adress.loc[df_adress.numro_dpartement == '2A', ['latitude', 'longitude']] = tmp.values
df_adress.loc[df_adress.numro_dpartement == '2A', 'numro_dpartement'] = '20'
df_adress = df_adress.loc[df_adress.numro_dpartement != '2B', :]

# dropna
df_adress = df_adress.dropna()

# as type float
df_adress.numro_dpartement = df_adress.numro_dpartement.astype(float)
```

```python
dico_adress = {}
depadress = df_adress.numro_dpartement.unique()

for dep in DEPARTEMENTS.keys():    
    if dep not in depadress:
        dico_adress[dep] = np.array([0, 0])
        print(dep, end=',')
    else:
        tmp = df_adress.loc[
             df_adress.numro_dpartement == dep,
             ['latitude', 'longitude']].values[0]
        tmp = [float(k.replace(',', '.')) if type(k) == str else k for k in tmp]
        
        dico_adress[dep] = np.array(tmp)
```

```python
tmp = df.groupby(['annais', 'preusuel', 'dpt']).nombre.sum().reset_index()

tmp['prenom_sans_accent'] = (tmp.preusuel.str.normalize('NFKD')
                            .str.encode('ascii', errors='ignore')
                            .str.decode('utf-8'))

tmp = tmp.loc[(tmp.prenom_sans_accent == 'JEAN')]
```

```python
years = sorted(tmp.annais.unique())

def f(i):
    
    print(years[i])
    tmpp = tmp.loc[(tmp.annais == years[i])]
    tmp2 = tmpp.dpt.apply(lambda x: dico_adress[x])
    bla = np.hstack((np.array([k for k in tmp2.values.flatten()]), np.array([tmpp.nombre.values]).T))
    
    fig, ax = plt.subplots(figsize=(6, 2))
    plt.title('Over time')
    tmp.groupby('annais').nombre.sum().plot(ax=ax)
    ax.axvline(years[i], ls='--', lw=3, c='gray')
    plt.show()
    
    centre = [46.227638, 2.213749]
    m = folium.Map(centre, zoom_start=6)
    HeatMap(bla, radius=40, blur=30).add_to(m)
    return m
```

```python
interact(f, i=(0, len(years) - 1));
```
